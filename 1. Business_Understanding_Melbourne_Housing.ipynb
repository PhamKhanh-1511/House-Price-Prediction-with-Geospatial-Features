# 📈 PHÂN TÍCH VÀ DỰ BÁO XU HƯỚNG GIÁ NHÀ TẠI MELBOURNE

---

## 1. Giới thiệu bài toán

Thị trường bất động sản từ lâu đã được xem là một trong những chỉ báo kinh tế vĩ mô quan trọng nhất, phản ánh sức khỏe tài chính nội tại, tốc độ đô thị hóa và xu hướng tích lũy tài sản của xã hội. Trong bối cảnh nền kinh tế số phát triển vượt bậc, việc quản lý và phân tích dữ liệu bất động sản không còn giới hạn ở các phương pháp thống kê truyền thống mà đã chuyển dịch mạnh mẽ sang hướng tiếp cận định lượng bằng **Khoa học dữ liệu (Data Science)**.

Thành phố Melbourne – trung tâm kinh tế và văn hóa lớn của Úc – sở hữu một thị trường nhà đất có quy mô giao dịch khổng lồ nhưng cũng mang đặc tính biến động rất phức tạp. Giá trị bất động sản tại đây chịu sự chi phối đan xen của nhiều nhóm yếu tố:

* **Yếu tố vị trí:** Khoảng cách địa lý di chuyển đến khu trung tâm thương mại (CBD), danh tiếng của từng khu vực ngoại ô (Suburbs).
* **Thông số kỹ thuật công trình:** Diện tích đất (`Landsize`), diện tích xây dựng (`BuildingArea`), số lượng phòng ngủ (`Rooms`/`Bedroom2`), phòng tắm (`Bathroom`) và vị trí đỗ xe (`Car`).

> **Thách thức:** Sự bất đối xứng thông tin và biên độ dao động giá lớn giữa các khu vực ngoại ô tạo ra những rào cản, rủi ro đáng kể cho các bên tham gia thị trường.

## 2. Mục tiêu phân tích & Câu hỏi nghiên cứu
Dự án được thực hiện nhằm khai thác bộ dữ liệu giao dịch bất động sản thực tế tại Melbourne (train.csv, giai đoạn 2016 - 2018). Mục tiêu cốt lõi là kiến tạo một lăng kính định lượng, giúp trả lời 3 câu hỏi nghiên cứu chiến lược (Research Questions - RQs) sau đây:

📌 RQ1: Market Dynamics & Spatial Differentiation (Động lực thị trường & Sự phân hóa không gian)

Nội dung phân tích: - Thị trường bất động sản Melbourne đang phân hóa như thế nào giữa các khu vực ngoại ô?

Đâu là những điểm nóng về thanh khoản và giá trị cao nhất?

Yếu tố khoảng cách đến CBD ảnh hưởng ra sao đến giá nhà?

Đặc trưng phân tích chính: Suburb, Distance, Price, Regionname

📌 RQ2: Sector Growth & Property Type (Tăng trưởng ngành & Loại hình bất động sản)

Nội dung phân tích:

Dựa trên giá trị trung bình và tốc độ tăng trưởng, đâu là loại hình bất động sản (nhà riêng, căn hộ, nhà phố) mang lại tiềm năng đầu tư hấp dẫn nhất?

Xu hướng biến động của các phân khúc giá (Thấp, Trung bình, Cao) theo thời gian ra sao?

Đặc trưng phân tích chính: Type (h, t, u), Phân vị Quantile, Date

📌 RQ3: Structural Features & Pricing Efficiency (Đặc trưng cấu trúc & Hiệu quả định giá)

Nội dung phân tích:

Các yếu tố cấu trúc công trình đóng góp như thế nào vào mô hình định giá?

Có thể xây dựng mô hình học máy phân loại chính xác phân khúc giá dựa trên các đặc trưng này hay không?

Đặc trưng phân tích chính: Rooms, Bathroom, Car, Landsize, BuildingArea

### 🎯 Đối tượng phục vụ của kết quả phân tích:

1. **Nhà đầu tư cá nhân & Quỹ đầu tư:** Xác định các khu vực và loại hình nhà có tiềm năng tăng giá tốt, tối ưu hóa danh mục đầu tư dài hạn.
2. **Đơn vị môi giới & Định giá:** Xây dựng công cụ định giá tự động, hỗ trợ tư vấn khách hàng chính xác, minh bạch.
3. **Cơ quan quản lý & Hoạch định chính sách:** Nắm bắt xu hướng đô thị hóa, phân bổ ngân sách hạ tầng phù hợp với các khu vực đang phát triển nóng.

## 3. Các yếu tố định hình giá bất động sản

Dựa trên bản chất của bộ dữ liệu thực tế, giá nhà tại Melbourne được định hình bởi 5 trụ cột thông tin trọng yếu:

* **Vị trí địa lý & Khoảng cách (`Distance`):** Các khu vực càng gần trung tâm thương mại (CBD) thường có giá đất cao hơn do tiện ích và thời gian di chuyển ngắn.
* **Quy mô & Cấu trúc (`Rooms`, `Bedroom2`, `Bathroom`, `Car`):** Số lượng phòng chức năng và chỗ đỗ xe phản ánh trực tiếp không gian sống và tiện nghi nội khu của tài sản.
* **Diện tích vật lý (`Landsize`, `BuildingArea`):** Đây là các chỉ số định lượng cốt lõi, mang tính chất quyết định trong bối cảnh quỹ đất đô thị ngày càng khan hiếm.
* **Tuổi thọ công trình (`YearBuilt`):** Công trình mới thường có giá cao hơn nhờ chất lượng kỹ thuật hiện đại; tuy nhiên các bất động sản cổ điển có giá trị kiến trúc lịch sử đặc biệt vẫn sở hữu mức giá rất cao.
* **Định danh khu vực (`Suburb`, `Postcode`):** Phản ánh uy tín, mức độ an ninh, hệ thống trường học công lập và các tiện ích công cộng của từng cộng đồng dân cư.

## 4. Cách tiếp cận phân tích (Methodology)

Dự án áp dụng quy trình phân tích và xử lý dữ liệu chuẩn mực bao gồm 6 giai đoạn:

```
[Tiền xử lý dữ liệu] ➔ [Truy vấn SQL & EDA] ➔ [Tính toán CAGR] ➔ [Huấn luyện ML] ➔ [Trực quan hóa] ➔ [Power BI Dashboard]

```

1. **Thu thập & Tiền xử lý dữ liệu (Data Preprocessing):** Xử lý dữ liệu khuyết thiếu, loại bỏ các bản ghi không hợp lý, mã hóa biến phân loại `Suburb` và gán nhãn phân khúc giá bằng phương pháp phân vị (Quantile).
2. **Khai phá dữ liệu & Truy vấn SQL (EDA & SQL Queries):** Sử dụng hệ quản trị **SQLite** kết hợp thư viện **Pandas** để truy xuất thống kê hiệu năng cao theo khu vực, thời gian và loại hình bất động sản.
3. **Mô hình hóa các chỉ số Tài chính (Financial Metrics):** Tính toán tốc độ tăng trưởng kép (**CAGR**) cho từng phân khúc giá để nhận diện động lực tăng trưởng bền vững của thị trường.
4. **Xây dựng mô hình học máy (Machine Learning Classification):** Triển khai các thuật toán Boosting mạnh mẽ (**XGBoost**, **LightGBM**, **Random Forest**) để phân loại tự động căn hộ vào 3 phân khúc giá (Thấp, Trung bình, Cao) dựa trên các đặc trưng đầu vào.
5. **Trực quan hóa dữ liệu (Data Visualization):** Sử dụng **Matplotlib** và **Seaborn** để xuất bản các biểu đồ trực quan (Biểu đồ nhiệt Heatmap, Scatter plot, Boxplot, Lục giác mật độ Hexbin và Biểu đồ đường xu hướng chuỗi thời gian).
6. **Xây dựng Dashboard tương tác (Power BI):** Chuyển hóa kết quả phân tích thành bảng điều khiển trực quan, cho phép người dùng tương tác lọc thông tin theo khu vực, loại hình nhà và dòng thời gian thực tế.

## 5. Mục tiêu kỳ vọng

* 📌 **Bằng chứng định lượng:** Cung cấp minh chứng rõ ràng về sự phân hóa giá nhà giữa các khu vực ngoại ô Melbourne, đặc biệt là ảnh hưởng mang tính quyết định của khoảng cách đến CBD.
* 📌 **Định hướng dòng tiền:** Nhận diện chính xác các loại hình bất động sản và phân khúc giá có tốc độ tăng trưởng kép (CAGR) dương ổn định, giúp tối ưu hóa hiệu quả đầu tư an toàn.
* 📌 **Độ chính xác mô hình:** Xây dựng thành công mô hình phân loại học máy với độ chính xác kỳ vọng **trên 75%**, hỗ trợ quy trình định giá và phân khúc thị trường hoàn toàn tự động.
* 📌 **Khung phân tích thực chiến:** Thiết lập một *Data-driven Framework* bền vững — tích hợp quy mô thị trường, cấu trúc công trình và yếu tố địa lý — làm cơ sở cho các quyết định đầu tư và phát triển bất động sản dài hạn trong tương lai.
